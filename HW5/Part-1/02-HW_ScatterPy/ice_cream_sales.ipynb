```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
temp = [14.2, 16.4, 11.9, 15.2, 18.5, 22.1, 19.4, 25.1, 23.4, 18.1, 22.6, 17.2]
sales = [215, 325, 185, 332, 406, 522, 412, 614, 544, 421, 445, 408]

#BONUS -Create a new list called "scoop_price", fill it with values, and then set it so that the size of the
#dots are set according to those values.

scoop_price = [120,100,80,60,40,20,10,30,50,70,90,110]
```

```python
# Tell matplotlib to create a scatter plot based upon the above data
plt.scatter(temp, sales, marker="o", facecolors="red", edgecolors="black",
            s=sales, alpha=1, sizes = scoop_price)
```

```python
# Set the upper and lower limits of our y axis
plt.ylim(min(sales), max(sales))
```

```python
# Set the upper and lower limits of our x axis
plt.xlim(min(temp)-1, max(temp)+1)
```

```python
# Create a title, x label, and y label for our chart
plt.title("Sales as a function of Temperature")
plt.xlabel("Temp (C)")
plt.ylabel("Sales ($)")
```
