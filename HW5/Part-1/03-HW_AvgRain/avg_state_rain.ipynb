```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load in csv
file = "Resources/avg_rain_state.csv"
df = pd.read_csv(file, low_memory = False)
df
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(df))
x_axis
tick_locations = x_axis
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize = (20,15))
plt.bar(x_axis, df["Inches"], color = 'red')
plt.xticks(tick_locations, df["State"], rotation = "vertical")
plt.yticks([0,25,50], [0,25,50])
```

```python
# Set x and y limits
plt.xlim(-.50, len(x_axis))
plt.ylim(0, max(df["Inches"])+10)
#plt.tight_layout()
```

```python
# Set a Title and labels
plt.title("Avg Rainfall per State")
plt.xlabel("State")
plt.ylabel("Inches Rainfall")
```

```python
# Save our graph and show the graph
plt.savefig("Images/rainfall.png")
plt.show()
```
